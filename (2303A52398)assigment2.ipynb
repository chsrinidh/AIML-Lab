```python
!pip install pandas
import pandas as pd
people = ['Kiran','Varun','Arun','Vijay']
age=[25,30,35,40]
height=[145,151,165,173]
weight=[45,55,65,75]
data={
    'Name':people,
    'Age': age,
    'Height': height,
    'Weight': weight
    }
df=pd.DataFrame(data)
print(df)
```

```python
import matplotlib.pyplot as plt
#3. Scatter Plot: Age vs Weight
plt.figure(figsize=(8, 6))
plt.scatter(df['Age'], df['Weight'], color='blue', marker='o')
plt.title('Scatter Plot of Age vs Weight')
plt.xlabel('Age (years)')
plt.ylabel('Weight (kg)')
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt
#3.2 bar Plot: Age vs Weight
plt.figure(figsize=(8, 6))
plt.bar(df['Age'], df['Weight'], color='blue')
plt.title('bar graph of Age vs Weight')
plt.xlabel('Age (years)')
plt.ylabel('Weight (kg)')
plt.show()
```
